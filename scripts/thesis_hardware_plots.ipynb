## SATSOP Alpha Course Day 2

```python
import numpy as np
import matplotlib.pyplot as plt

filename = "/home/andrew/bags/thesis/SATSOP_Alpha_course_day2_map.csv"
voxel_size = 0.2
data = np.array(np.genfromtxt(filename, delimiter=",", skip_header=2))
t = np.arange(0, 5*(len(data[:,0])), 5)
fig, ax = plt.subplots(figsize=[12.0, 7.0])
ax.tick_params(axis='both', which='major', labelsize='16')
plt.plot(t, data[:,1]*voxel_size**3)
plt.xlabel("Exploration Time [s]", fontsize='16')
plt.ylabel("Map Volume [m^3]", fontsize='16')
```
