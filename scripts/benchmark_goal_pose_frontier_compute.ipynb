## Benchmark IFVE components

```python
import numpy as np
import matplotlib.pyplot as plt
import csv

filename = "/home/andrew/Desktop/octomaps/goal_pose_sampling_and_frontier_benchmark_urban.csv"
data_urban = np.genfromtxt(filename, dtype=int, delimiter=",", names=True)
labels = data_urban.dtype.names

ind = range(1, len(data_urban)+1)
fig, ax = plt.subplots(figsize=[12.0, 7.0])
ax.tick_params(axis='both', which='major', labelsize='16')
for i in range(1, len(labels)):
    if i == 1:
        ax.bar(data_urban[labels[0]]/1e6, data_urban[labels[i]], label=labels[i], width=0.1)
        previous = data_urban[labels[i]]
    else:
        ax.bar(data_urban[labels[0]]/1e6, data_urban[labels[i]], label=labels[i], width=0.1, bottom=previous)
        previous = previous + data_urban[labels[i]]
ax.set_ylabel('Compute (ms)', fontsize="16")
ax.set_xlabel('Voxels (millions)', fontsize="16")
# ax.set_title('IFVE Urban Benchmark Compute')
ax.legend(fontsize="16")
```
